```python
import os 

import pandas as pd 
```

```python
input_dirs = [
    "/Users/jimuelcelestejr/Documents/codebook/MLSpeech4MH/results/TAUKADIAL2024_samples/acoustic/egemaps_functionals",
    "/Users/jimuelcelestejr/Documents/codebook/MLSpeech4MH/results/TAUKADIAL2024_samples/acoustic/boaw_size_100_compare_2016_lld"
]

file = "taukdial-003-2.csv"
input_files = []
for input_dir in input_dirs:
    input_file = os.path.join(input_dir, file)
    input_files.append(input_file)
    print(input_file)
```

```python
def merge_features(input_files):
	features_list = []
	
	for input_file in input_files:
		print(input_file)
		features = pd.read_csv(input_file)
		features_list.append(features)
		print(features) 

	print(features_list)
	concatenated_features = pd.concat(features_list, axis='columns')

	return concatenated_features

merged_features = merge_features(input_files)
merged_features
```

```python
output_dir = "/Users/jimuelcelestejr/Documents/codebook/MLSpeech4MH/results/sandbox"
output_file = os.path.join(output_dir, file)
print(output_file) 
merged_features.to_csv(output_file, index=False)
```
