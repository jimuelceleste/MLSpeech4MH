```python
import os

import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
egemaps_dir = "/Users/jimuelcelestejr/Documents/codebook/MLSpeech4MH/results/TAUKADIAL2024/machine_learning/egemaps_experiments"

models = [
    "svm_classifier",
    "logistic_regression_classifier", 
    "xgb_classifier", 
    "random_forest_classifier"
]
```

```python
for model in models:
    perf_file = os.path.join(egemaps_dir, model, "performance.csv") 
    perf_df = pd.read_csv(perf_file, index_col=0) 
    print(model) 
    print(perf_df)
```

```python
sns.boxplot(perf_df.iloc[0:5]["f1_score"])
```

```python
sns.set_context("poster")
sns.set_style('white')
sns.set_palette("viridis")
plt.figure(figsize=(12, 8))
plt.title("Extended Geneva Minimalistic Acoustic Feature Set", fontsize=30)
plt.xlabel("Metrics", fontsize=24)
plt.ylabel("Performance", fontsize=24)
sns.despine()
# plt.rc('font', size=12)
# plt.legend(fontsize=20)
plt.ylim(0.0, 1.0)
sns.boxplot(perf_df.iloc[0:5], width=0.5)
```
